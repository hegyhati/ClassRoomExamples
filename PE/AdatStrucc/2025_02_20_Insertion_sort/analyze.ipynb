```python
results = {
    "increasing" : {},
    "decreasing" : {},
    "random_1" : {},
    "random_2" : {},
    "random_3" : {},
    "random_4" : {},
    "random_5" : {},
    "random" : {},
}

def get_size_and_case(filename:str) -> tuple[int,str]:
    tmp = filename.removeprefix("examples/")
    tmp = tmp.removesuffix(".txt")
    return int(tmp[:10]), tmp[11:]

with open("allnighter.csv") as f:
    for result in f:
        result = result.split(",")
        size,case = get_size_and_case(result[0])
        results[case][size] = int(result[1])

for size, ticks in results["random_1"].items():
    results["random"][size] = sum(results[f"random_{i}"][size] for i in range(1,6)) // 5
```

```python
import matplotlib.pyplot as plt

for case in ["increasing", "decreasing", "random"]:
    plt.plot(results[case].keys(),results[case].values(), label = case)
plt.legend()
```
