```python
import pandas as pd
pd.options.mode.copy_on_write = True
```

```python
data = pd.read_csv("mount_everest_deaths.csv")
data
```

```python
data["year"] = data.Date.str[-4:]
years = data.year.unique()
years
```

```python
bad_indices = data[data.year == "g-75"].index
data.drop(bad_indices, inplace=True)
data
```

```python
years = data.year.unique()
years
```

```python
data.year = data.year.astype(int)
data.year
```

```python
data.Age.unique()
```

```python
data.dropna(subset=["Age"], inplace=True)
data.Age.unique()
```

```python
data.Age = data.Age.astype(int)
data.Age
```

```python
data.plot(x="year", y="Age", kind="scatter")
```

```python
by_country = data.groupby("Nationality").count()
by_country.sort_values("Name", ascending=False, inplace=True)
by_country.plot(y="Name", kind="bar", legend=False)
```

```python
data.Nationality.value_counts().plot(kind="bar")
```

```python
from_nepal = data[data.Nationality == "Nepal"]
from_nepal["Cause of death"].value_counts().plot(kind="bar")
```

```python
from_nepal["Cause of death"].unique()
```

```python
to_replace = {
    "Heart Attack": "Heart attack",
    "Stroke": "Heart attack",
    "by fall" : "Fall",
    "Disappearance (likely accidental death during climb)": "Disappearance",
    "Fall into a crevasse": "Falling into a crevasse",
    "2014 Mount Everest Avalanche" : "Avalanche",
    "Base Camp avalanche following the April 2015 Nepal earthquake": "Avalanche",
    "Disappearance (reportedly suffered from snow blindness)" : "Disappearance"
}
for key,value in to_replace.items():
    from_nepal.replace(key, value, inplace=True)

from_nepal["Cause of death"].value_counts().plot(kind="bar")
```
