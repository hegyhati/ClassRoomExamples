```python
import pandas as pd
```

```python
df = pd.read_csv("okcupid_profiles.csv")
list(df)
```

```python
df.dropna(subset=["age", "status", "sex", "height", "job", "religion"], inplace=True)
df
```

```python
df.plot(kind="scatter", x="age", y="height")
```

```python
woman = df[df.sex=="f"]
man = df[df.sex=="m"]
woman
```

```python
man.income.value_counts().plot(kind="bar")
```

```python
woman.income.value_counts().plot(kind="bar")
```

```python
man_with_income = man[man.income != -1]
man_with_income.income.value_counts().plot(kind="bar")
```
